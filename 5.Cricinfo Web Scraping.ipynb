```python
from requests import get
from bs4 import BeautifulSoup
```

```python
url = 'http://www.espncricinfo.com/rankings/content/page/211271.html'
```

```python
response = get(url)
response.status_code
content = response.content
```

```python
parser = BeautifulSoup(content, "html.parser")
```

```python
cric_format = parser.find_all("div",class_= "ciPhotoContainer")[0].select("h3")
cric_format
```

```python
format_ranking = parser.find_all("table", class_="StoryengineTable")
# format_ranking
```

```python
for i, value in enumerate(format_ranking):
    print(cric_format[i].text)
    print(value.caption.text)
    print()
    tr = value.find_all("tr")
    i = 0
    for t in tr:
        if i ==0:
            search = "th"
        else:
            search = "td"
        for c in t.findChildren(search):
            print("{:20}".format(c.text), end = "")
        print()
        i += 1
    print()
        
```
